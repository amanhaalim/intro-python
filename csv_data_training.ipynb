## 🧹 Section 4: Cleaning the Data

Real-world data is almost always messy. Common issues include:
- ❌ Missing values (NaN)
- 🔁 Duplicate rows
- ❗ Wrong data types
- 🔡 Inconsistent formatting

### 4.1 — Finding Missing Values

````python
# Check for missing values
print('🔍 Missing Values per Column:')
missing = df.isnull().sum()
print(missing[missing > 0])

# Total percentage of missing data
total_cells = df.shape[0] * df.shape[1]
missing_cells = df.isnull().sum().sum()
print(f'\n📉 Total missing: {missing_cells} / {total_cells} cells ({missing_cells/total_cells*100:.2f}%)')
````

````python
# Visualise missing values as a heatmap
plt.figure(figsize=(12, 4))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Value Heatmap (Yellow = Missing)', fontsize=14)
plt.tight_layout()
plt.show()
````

### 4.2 — Handling Missing Values

````python
# Strategy: Fill missing MathScore with the column median
median_math = df['MathScore'].median()
print(f'📐 Median Math Score: {median_math}')

df['MathScore'] = df['MathScore'].fillna(median_math)

# Verify no more missing values
print(f'Missing in MathScore after filling: {df["MathScore"].isnull().sum()}')
````

````python
# Confirm all missing values are handled
print('✅ Remaining Missing Values:')
print(df.isnull().sum())
````

### 4.3 — Checking for Duplicate Rows

````python
# Check for duplicates
duplicate_count = df.duplicated().sum()
print(f'🔁 Duplicate rows found: {duplicate_count}')

# Drop duplicates if any
df = df.drop_duplicates()
print(f'✅ Shape after removing duplicates: {df.shape}')
````

### 4.4 — Fixing Data Types & Formatting

````python
# Strip whitespace from string columns
str_cols = df.select_dtypes(include='object').columns
df[str_cols] = df[str_cols].apply(lambda x: x.str.strip())

# Standardise 'FeesPaid' to boolean
df['FeesPaid'] = df['FeesPaid'].map({'Yes': True, 'No': False})

# Compute an average score column
df['AvgScore'] = df[['MathScore', 'ScienceScore', 'EnglishScore']].mean(axis=1).round(2)

print('✅ Data types after cleaning:')
print(df.dtypes)
df.head()
````

### 4.5 — Filtering & Selecting Data

````python
# Filter students with Grade A
top_students = df[df['Grade'] == 'A']
print(f'🏅 Number of Grade-A students: {len(top_students)}')
top_students[['Name', 'City', 'AvgScore', 'Attendance(%)']]
````

````python
# Filter students who haven't paid fees and have low attendance
at_risk = df[(df['FeesPaid'] == False) & (df['Attendance(%)'] < 85)]
print(f'⚠️ At-risk students (unpaid fees + low attendance): {len(at_risk)}')
at_risk[['Name', 'City', 'Attendance(%)', 'Grade', 'FeesPaid']]
````

## 📊 Section 5: Visualising the Data

Visualisations help us spot trends, distributions, and outliers at a glance.

### 5.1 — Score Distributions (Histogram)

````python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

subjects = ['MathScore', 'ScienceScore', 'EnglishScore']
colors = ['steelblue', 'seagreen', 'coral']

for ax, subject, color in zip(axes, subjects, colors):
    ax.hist(df[subject], bins=10, color=color, edgecolor='white', alpha=0.85)
    ax.set_title(f'{subject} Distribution', fontsize=13, fontweight='bold')
    ax.set_xlabel('Score')
    ax.set_ylabel('Number of Students')
    ax.axvline(df[subject].mean(), color='red', linestyle='--', label=f'Mean: {df[subject].mean():.1f}')
    ax.legend()

plt.suptitle('Score Distributions Across Subjects', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

### 5.2 — Gender Distribution (Pie Chart)

````python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    colors=['#4C72B0', '#DD8452'],
    startangle=140,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)
plt.title('Gender Distribution of Students', fontsize=14, fontweight='bold')
plt.show()
````

### 5.3 — Grade Distribution (Count Plot)

````python
plt.figure(figsize=(8, 5))
grade_order = ['A', 'B', 'C']
palette = {'A': '#2ecc71', 'B': '#3498db', 'C': '#e74c3c'}

sns.countplot(data=df, x='Grade', order=grade_order, palette=palette, edgecolor='white')
plt.title('Grade Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
````

### 5.4 — Average Score by Gender (Box Plot)

````python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Gender', y='AvgScore', palette='Set2')
plt.title('Average Score Distribution by Gender', fontsize=14, fontweight='bold')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.tight_layout()
plt.show()
````

### 5.5 — Attendance vs Average Score (Scatter Plot)

````python
plt.figure(figsize=(9, 6))
palette = {'A': '#2ecc71', 'B': '#3498db', 'C': '#e74c3c'}

for grade, group in df.groupby('Grade'):
    plt.scatter(
        group['Attendance(%)'],
        group['AvgScore'],
        label=f'Grade {grade}',
        color=palette[grade],
        s=80, alpha=0.85, edgecolors='white'
    )

# Add a trend line
z = np.polyfit(df['Attendance(%)'], df['AvgScore'], 1)
p = np.poly1d(z)
x_line = np.linspace(df['Attendance(%)'].min(), df['Attendance(%)'].max(), 100)
plt.plot(x_line, p(x_line), 'k--', linewidth=1.5, label='Trend Line')

plt.title('Attendance vs Average Score', fontsize=14, fontweight='bold')
plt.xlabel('Attendance (%)')
plt.ylabel('Average Score')
plt.legend()
plt.tight_layout()
plt.show()
````

### 5.6 — Correlation Heatmap

````python
numeric_cols = ['Age', 'MathScore', 'ScienceScore', 'EnglishScore', 'Attendance(%)', 'AvgScore']
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(9, 7))
sns.heatmap(
    corr_matrix,
    annot=True, fmt='.2f',
    cmap='coolwarm',
    linewidths=0.5,
    square=True
)
plt.title('Correlation Heatmap of Numeric Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

### 5.7 — Top 10 Students by Average Score (Bar Chart)

````python
top10 = df.nlargest(10, 'AvgScore')[['Name', 'AvgScore', 'Grade']].reset_index(drop=True)

plt.figure(figsize=(11, 5))
bars = plt.barh(top10['Name'], top10['AvgScore'], color='steelblue', edgecolor='white')

for bar, score in zip(bars, top10['AvgScore']):
    plt.text(bar.get_width() - 1.5, bar.get_y() + bar.get_height()/2,
             f'{score}', va='center', ha='right', color='white', fontweight='bold')

plt.gca().invert_yaxis()
plt.title('Top 10 Students by Average Score', fontsize=14, fontweight='bold')
plt.xlabel('Average Score')
plt.tight_layout()
plt.show()
````

## 📌 Quick Reference — Pandas Cheat Sheet

| Task | Code |
|---|---|
| Load CSV | `pd.read_csv('file.csv')` |
| View top rows | `df.head(n)` |
| Shape | `df.shape` |
| Info | `df.info()` |
| Missing values | `df.isnull().sum()` |
| Fill missing | `df['col'].fillna(value)` |
| Drop duplicates | `df.drop_duplicates()` |
| Filter rows | `df[df['col'] > value]` |
| Select columns | `df[['col1', 'col2']]` |
| Group by | `df.groupby('col').mean()` |
| Save CSV | `df.to_csv('file.csv', index=False)` |
